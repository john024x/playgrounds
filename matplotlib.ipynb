```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y,'rD:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show
```

```python
plt.boxplot(x)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.plot(y,x)
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1.5])
axes.plot(x,y,'b')
fig.show()
```

```python
z = np.sin(x)
fig,axes = plt.subplots()
axes.plot(x,z)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,z)
axes[1].plot(z,y)
```

```python
fig,axes = plt.subplots(2,4,figsize=(15,15))
axes[0,0].plot(x,np.cos(x))
axes[0,0].set_title('Cos X') #Establecer un titulo para la grafica
axes[0,0].set_xlabel('X') #Establecer un label en X
axes[0,0].set_ylabel('Y') #establecer un label en Y
axes[0,1].plot(x,np.tan(x))
axes[0,1].set_title('Tan X')
axes[0,2].plot(x,np.sin(x))
axes[0,2].set_title('Sin X')
axes[0,3].plot(x,np.cos(x)*-1)
axes[1,0].plot(x,np.tan(x)*-1)
axes[1,1].plot(x,np.sin(x)*-1)
plt.legend(loc='lower left', bbox_to_anchor=(1,0.2))
plt.style.use('dark_background')
```
