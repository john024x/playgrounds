--- Copy

```python
array = np.arange(0,11)
array
```

```python
trozo_arr = array[0:6]
```

```python
trozo_arr
```

```python
trozo_arr[:] = 0
print(trozo_arr)
print(array)
```

```python
array_copy = array.copy()
```

```python
array_copy[:] = 0
print(array_copy)
print(array)
```
