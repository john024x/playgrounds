```python
#Mapea los datos del CSV mediante un delimitador linea por linea, separados por coma
equiposOlimpicosRDD = sc.textFile(PATH+"paises.csv").map(lambda line: line.split(","))

#al ejecutar la accion es cuando toma sentido la transoformacion previa, si no hubiese existido el archivo, sería en este punto cuando arrojaría el error, no antes
equiposOlimpicosRDD.take(25) 
```

```python
#Hacemos una funcion almbda que nos de la segunda columna, filtramos por los datos distintos entre si, y hacemos un count para saber los paises unicos de la competencia
equiposOlimpicosRDD.map(lambda x:(x[2])).distinct().count()
```

```python
#Se hace un mapeo de los parametros, mediante la funcion lambda los acomodamos llave valor segun el pais y los equipos, posteriormente los agrupamos por llave
#El mapValues, es un mapeo sobre el valor de cada llave para acomodarlos en forma de lista
equiposOlimpicosRDD.map(lambda x: (x[2],x[1])).groupByKey().mapValues(list).take(15) #Take toma unicamente algunos registros, y no todos como lo hace collect
```

```python
Argentina = equiposOlimpicosRDD.filter(lambda l: "ARG" in l)#.groupByKey().mapValues(list).take(15)
Argentina.collect() #Collect manda todos los RDD que tengan el valor deseado a la computadora, por lo que puede generar problemas de memoria
```

```python
resultado = sc.textFile(PATH+"resultados.csv").map(lambda l:l.split(","))
```

```python
resultado = resultado.filter(lambda l: 'NA' not in l[1]) #En este paso estamos filtrando los resultados que no tengan medallas
```

```python
resultado.take(10)
```

```python
DeportistaPaises = deportistaOlimpicoRDD.map(lambda l: [l[-1],l[:-1]]).join(equiposOlimpicosRDD.map(lambda x:[x[0],x[2]]))
```

```python
resultadoFinal = resultado.map(lambda l : [l[2],l[:-1]]).join(DeportistaPaises.map(lambda x : [x[1][0][0],x[1]]))

resultadoFinal.take(10)
```

```python
valoresMedalla = {'Gold': 7, 'Silver': 5, 'Bronze': 4}
paisesMedalla = DeportistaPaises.join(resultado)
paisesMedalla = paisesMedalla.map(lambda x: [x[1][0][-1],valoresMedalla[x[1][1]]])
#al incluir como segundo parametro de la lambda function, el diccionario de valores medalla, nos da en lugar de la medalla, el valor de la misma
```

```python
from operator import add #Importamos para poder hacer la suma

#Hace un reduce por su llave, acumulando los valores mediante su llave de pais, y valor de el acumulado, por orden descendente
conclusion = paisesMedalla.reduceByKey((add)).sortBy(lambda x:x[1],ascending=False)
conclusion.take(10) 
```

#Parte referente a DF desde un RDD

```python
def eliminarEncabezado(index,iterator): #Elimina el encabezado
    return iter(list(iterator)[1:])
```

```python
deportistaOlimpicoRDD = deportistaOlimpicoRDD.mapPartitionsWithIndex(eliminarEncabezado)
deportistaOlimpicoRDD.take(5)
```

```python
deportistaOlimpicoRDD = deportistaOlimpicoRDD.map(lambda l:( # Se define la estructura para lo que seran las columnas 
    int(l[0]),
    l[1],
    int(l[2]),
    int(l[3]),
    int(l[4]),
    float(l[5]),
    int(l[6])
))
```

```python
from pyspark.sql import SQLContext
from pyspark.sql.types import StructType, StructField, IntegerType,StringType, FloatType,Row
sqlContext = SQLContext(sc)
```

```python
schema = StructType([ #definimos la estructura del schema para el RD
    StructField("deportista_id",IntegerType(),False),
    StructField("nombre",StringType(),False),
    StructField("genero",IntegerType(),False),
    StructField("edad",IntegerType(),False),
    StructField("altura",IntegerType(),False),
    StructField("peso",FloatType(),False),
    StructField("equipo_id",IntegerType(),False),
])
```

```python
deportistaDF = sqlContext.createDataFrame(deportistaOlimpicoRDD,schema)
```
